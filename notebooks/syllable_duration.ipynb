```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pyfinch.database.load import ProjectLoader
from pyfinch.utils.draw import remove_right_top
from pyfinch.utils import save
import seaborn as sns
import sys
sys.path.append(r"..")
```

```python
# data_path = r"H:\Box\Data\Deafening Project\Analysis\SyllableDuration"
data = ProjectLoader().path / 'Analysis/SyllableDuration/SyllableDuration.csv'
```

```python
df = pd.read_csv(data)
```

```python
df.columns
```

```python
df = df.query("Context == 'U' and TaskName == 'Predeafening'")
```

```python
fig, ax = plt.subplots(figsize=(9,6))
sns.histplot(x="Duration",
             data=df,
             bins=100,
             hue="SyllableType",
             kde=True, ax=ax)
ax.set_xlim([0, 350])
ax.set_ylim([0, 1300])
remove_right_top(ax)
plt.show()
```
